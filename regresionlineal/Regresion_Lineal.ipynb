### 1. Introducción

El modelo de regresión lineal consiste en la estimación de una función lineal que relacione una variable dependiente $ y $ a partir de unas variables independientes o predictoras ($X_{1},X_{2},...X_{n} $). Dicha función puede estar representada así:

$$
y = w_{i}X_{i} + b
$$

En general, este modelo tiene diversas aplicaciones y es una de las formas más sencillas en el aprendizaje de máquinas, siendo un muy buen punto de partida para el aprendizaje y entendimiento de la inteligencia artificial.

Para poder encontrar los valores de $ w_{i} $, existen diferentes estrategias como los Mínimos Cuadrados Ordinarios (MCO) o métodos iterativos más recientes como el Gradiente Descendiente. En ambos casos, se busca minimizar los errores de la estimación.

### 2. Historia

El origen de la noción de regresión proviene, en gran parte, de áreas del conocimiento como Biología y Eugenesia (Castro, Gea, Cañadas & Contreras, 2012) [1] . El primer autor que se interesa en el tema fue LambertAdolphe-Jacques Quetelet quien en sus estudios relaciona dos o más variables, por ejemplo llega a obtener una ecuación de una hipérbola que relaciona la edad y la altura de las personas entre cero y 30 años.

Sin embargo, el término regresión fue introducido por Galton en su libro “Natural inheritance” (1889) refiriéndose a la “ley de la regresión universal”; su trabajo se centraba en la descripción de los rasgos físicos de los descendientes (una variable) a partir de los de sus padres (otra variable) (Contreras & del Pino, n.d.) [2]

Las ideas modernas sobre regresión se originan en los trabajo de Legendre y Gauss, sobre el método de mínimos cuadrados, para ajustar los datos sobre las órbitas de cuerpos celestes. El primer estudio documentado sobre el método de mínimos cuadrados, de donde deriva la idea de regresión, es debido a Legendre (1752-1833) en 1805.

### 3. El algoritmo de regresión lineal

El modelo de regresión lineal puede ser un modelo simple, en el caso en elq ue solo tenemos una variable predictora o múltiple si tenemos 2 o más de estas variables. Para generalizar, se puede decir que el modelo de regresión lineal puede representarse con la siguiente ecuación:

$$
Y =w_0 +  w_1x_1 + w_2x_2  + ... +  \epsilon
$$

donde, 

$Y:$ representa a la variable independiente.

$w_i:$ representa los pesos de las variables predictoras o explicativas.

$x_i:$ representa las variables represoras o predictoras.

$\epsilon:$ un término de error aleatorio que corresponde a los factores no medibles por las variables $x_i$.

El modelo de regresión lineal múltiple puede ser representado en forma matricial y la solución para encontrar los pesos de $w_i$ se puede representar como la siguiente ecuación matricial:

$$
w_i =  (X^TX)^{-1}X^TY
$$

#### 4.2. Preprocesamiento, limpieza, transformación y visualización inicial



Ahora bien, dentro del proceso de preprocesamiento, es importante validar que los valores de las variables object sean consistentes y estén estandarizados. Un sólo caracter diferente como una tílde, un espacio o signo de puntualción pueden cambiar su valor. Vamos a revisar esto.

```python
##Variables tipo object para evidencias sus valores unicos

##Seleccionar solo las columnas de tipo objet = 'O'
listobj =[]
columns = db.columns

#ingresar a la lista unicamente los nombres de las columnas con tipo object
for variable in columns:
    if db[variable].dtype=='O':
        listobj.append(variable)

for variable in listobj:
    valores_unicos=db[variable].unique()
    print(valores_unicos)
```

Todas las variables tienen valores únicos (variables tipo Object). Podemos continuar con nuestro análisis.

```python
##Missing values
#Debemos revisar el porcentaje de los missing en cada variable. Creamos una función que retorne una tabla con las variables, y porcentajes mayores a cero.
def tabla_missing(db):
    is_null = db.isnull().sum().sort_values(ascending=False) #Validar si una variable tiene missing en alguno de sus valores y sumarlos (total missing)
    porc = ((db.isnull().sum()/db.isnull().count()).sort_values(ascending=False))*100

    missing_data = pd.concat([is_null, round(porc,2)], axis=1, keys=['Total', 'Percent'])
    return missing_data[missing_data.Percent>0] #Imprimir sólo las variables con porcentaje de missing mayores a cero
```

```python
#Invocar la función anteriormente creada
tabla_missing(db)
```

```python
#Realizamos una imputación de datos con el promedio para las variables normalized_losses, stroke, bore.

db['normalized_losses'].fillna(db['normalized_losses'].mean(), inplace=True)
db['stroke'].fillna(db['stroke'].mean(), inplace=True)
db['bore'].fillna(db['bore'].mean(), inplace=True)

#eliminamos las demás registros con missing de las otras variables:

db.dropna(subset=['price'], axis=0, inplace=True)
db.dropna(subset=['horsepower'], axis=0, inplace=True)
db.dropna(subset=['peak_rpm'], axis=0, inplace=True)
db.dropna(subset=['num_of_doors'], axis=0, inplace=True)
```

```python
#Validar que no haya variables con missing 
tabla_missing(db)
```

##### 4.3.2 Modelo de regresión lineal simple (una variable predictora) con sklearn

Para este caso y de acuerdo con el gráfico de 4.3.1.3, se va a seleccionar la variable engine_size para realizar las predicciones de los precios de los carros en este primer modelo lineal simple.

```python
#Separar la variable para el modelo en 4 nuevas variables: x1_train, y_1_train, x_1_test, y_1_test
x_1_train = db_train['engine_size']
y_1_train = db_train['price']

x_1_test = db_test['engine_size']
y_1_test = db_test['price']
```

```python
#Graficamos la variable predictora y la variable de interés para ver su relación:
```

```python
#Creamos la gráfica
def relation_engine_price(es, price):
  plt.scatter(es, price, alpha=0.25)
  plt.title("Cilindraje contra Precio de carros")
  plt.show()
  return np.corrcoef(es, price)[0, 1]   #  se recibe uma matriz de correlaciones. Se xtrae la correlación
```

```python
print("La correlación de Cilindraje y Precio de carros es:", relation_engine_price(x_1_train, y_1_train))
```

```python
#Convertir la variable en un arreglo de 1D (una dimension)
x_1_train = np.array(x_1_train)
x_1_train = x_1_train.reshape(-1,1)

y_1_train = np.array(y_1_train)
y_1_train = y_1_train.reshape(-1,1)

x_1_test = np.array(x_1_test)
x_1_test = x_1_test.reshape(-1,1)

y_1_test = np.array(y_1_test)
y_1_test = y_1_test.reshape(-1,1)

```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(x_1_train,y_1_train)

print(lm.intercept_)
print(lm.coef_)
```

```python
#Graficar la recta de regresión
y_1_predict = lm.predict(x_1_test)

plt.plot(x_1_test, y_1_predict, color="green")
plt.scatter(x_1_test, y_1_test, alpha=0.25)
plt.show()

from sklearn.metrics import mean_squared_error

fit_intercept_error = mean_squared_error(y_1_test, y_1_predict)

w1= lm.coef_[0]
w0 = lm.intercept_
result_score =lm.score(x_1_test,y_1_test)

print("Modelo Cilindraje vs precio: w1 =", round(w1[0],3), "w0 =",
        round(w0[0],3), "mse =", round(fit_intercept_error,3), "score =", round(result_score,3) )

```
